```python
import numpy as np
#X=(Area,balcony,num_of_malls,distance_from_metro, num_of_hours_of_electricity_supply,num_of_flats, num_of_schools)
X=np.array(([2000,200,3,500,24,100,5],[3000,500,1,5000,18,1100,1],[1200,50,4,100,22,800,3],[1500,200,2,200,21,500,3]),dtype=float)
X=X/np.amax(X,axis=0)#Scaling va;ue of each feature by dividing it with max value of each feature
#Y=output vector
Y=np.array(([100],[70],[60],[80]),dtype=float)
Y=Y/100.0#Scaling value of output
print(Y)
```

```python
class Neural_Network(object):
    def __init__(self): #deciding parameters of neural network
        self.inputSize=7
        self.outputSize=1
        self.hiddenSize=5 #hidden size is no of hidden nodes in single hidden layer
        #now randomly generating weights
        self.W1=np.random.randn(self.inputSize,self.hiddenSize)
        self.W2=np.random.randn(self.hiddenSize,self.outputSize)
    def forward(self,X):#forward propogation function
        self.Z=np.dot(X,self.W1) #dot product is matrix multiplication
        self.Z=self.sigmoid(self.Z)#Applying activation function
        self.O=np.dot(self.Z,self.W2)
        self.O=self.sigmoid(self.O)
        return self.O
    def sigmoid(self,s):
        #activation function
        return 1/(1+np.exp(-s))
    #for backpropogation
    def sigmoidPrime(self,s): #rate of change of activation funcion
        return s*(1-s)
    def backpropogation(self, X,Y,O):
        self.o_error=Y-O #finding margin of error of output layer
        #The error computed is not mean squared sum error
        self.o_delta=self.o_error*self.sigmoidPrime(O) #applying derivative of sigmoid to error
        print(self.sigmoidPrime(O).shape)
        self.z2_error=self.o_delta.dot(self.W2.T)  #how much hidden layer output contributed to error
        self.z2_delta=self.z2_error*self.sigmoidPrime(self.Z) #applying derivative of sigmoid to z2 error
        
        self.W1=self.W1+X.T.dot(self.z2_delta)#adjusting first set of weights
        self.W2=self.W2+self.Z.T.dot(self.o_delta)
        
    def train(self,X,Y):
        o=self.forward(X)
        self.backpropogation(X,Y,o)
```

```python
NN=Neural_Network()
for i in range(100):#train the NN 100 times
    NN.train(X, Y)
print("Actual Value\n"+str(Y))
print("Predicted Value\n"+str(NN.O))

```
