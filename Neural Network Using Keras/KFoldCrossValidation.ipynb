# Setting Up Environment

```python
#In K fold validation the test data is divided into K groups and all except one group is used for training. 
#Every group becomes test group for once and only once
#The average loss of all the k scenarios will be considered as final loss
from keras.models import Sequential
from keras.layers import Dense #each neuron in dense layer recieves input from every neuron from previous layer
from sklearn.model_selection import StratifiedKFold
import numpy as np
seed=10
np.random.seed(seed)
```

# 1. Loading Data 

```python
data=np.loadtxt("pima-indians-diabetes.csv",delimiter=",")
X=data[:, 0:8]
Y=data[:, 8]
```

# Defining Model, Compile, Fit, Evaluate

```python
kfold=StratifiedKFold(n_splits=10, shuffle=True, random_state=seed) #Creating an object of StratifiedKFold
score=[]
for train, test in kfold.split(X,Y):
    print(train)
    model=Sequential()
    #We do not have to explicitly add input layer
    #We only add hidden layers and output layers
    model.add(Dense(12,input_dim=8,activation='relu'))#Add Dense (Fully connected layer with 12 neurons, input of 8, activation relu)
    model.add(Dense(8,input_dim=12,activation='relu'))
    model.add(Dense(1,input_dim=8,activation='sigmoid'))
     #Model actually comes into existense; ready for training and later prediction
    model.compile(loss='binary_crossentropy',optimizer="adam",metrics=['accuracy'])#optimizer is used for updating the weights 
    #e.g., gradient descent or momentum or rms prop or adam
    #for classification problems, accuracy is default metric
    #fitting the model means TRAINING the model using train data and selecting the best parameters
    #epochs means number of iterations over entire x and y data provided for training model
    #batch_size specifies number of instances of training input that are evaluated before weight update actually happens
    #epochs and batch size must be chosen by hit and trial
    #validation_split splits the dataset into training data and test data
    model.fit(X[train],Y[train],epochs=150,batch_size=10,verbose=0)
    #The output will display validation accuracy after each epoch
    scores=model.evaluate(X[test],Y[test])
    print("\n %s %f\n" %(model.metrics_names[1], scores[1]*100))
    score.append(scores[1]*100)
print("%f %% %f %%" %(np.mean(score), np.std(score)))
```
