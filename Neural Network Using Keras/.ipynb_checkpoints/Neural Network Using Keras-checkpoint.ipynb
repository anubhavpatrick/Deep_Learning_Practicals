# Importing necessary library and packages

```python
from keras.models import Sequential #for ANN
from keras.layers import Dense #for hidden layers
import numpy as np
```

```python
np.random.seed(5) #setting random seed  for reproducability of same results
```

# 1. Load Data

```python
#load the data set which contains 650 records about whether a website is legitimate or not
dataset = np.loadtxt("phishingData.csv",delimiter=",")
#Converting Data into training and testing
X_Train=dataset[1:501,0:10]
Y_Train=dataset[1:501:,10]
X_Test=dataset[501:  , 0:10]
Y_Test=dataset[501:  , 10]
```

# 2. Define Model

```python
#we will use sequential model
model=Sequential()
model.add(Dense(12,input_dim=10,activation='relu')) #Fully connected layers are defined using Dense
model.add(Dense(8,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

# 3. Compile Model

```python

import tensorflow
import keras
#print(tensorflow.__version__)
#print(keras.__version__)
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

# 4. Fit the model

```python
# Fit the model
model.fit(X_Train, Y_Train, epochs=150, batch_size=10)
```

# Evaluate Model

```python
# evaluate the model
scores = model.evaluate(X_Test, Y_Test)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
